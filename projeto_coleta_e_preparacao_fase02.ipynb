## Correções nos Dados Referentes ao Ano de 2016





Na **coluna Q8** do DataFrame de 2016, os dados estão em formato de ***string***, enquanto nos anos seguintes, a formatação foi alterada para ***inteiros***, com valores variando de 0 a 10.

```python
df_2016['Q8'].value_counts()
```

Para realizar a conversão de string para inteiro, foram definidos os seguintes parâmetros, variando de 0 a 10.

|String|Inteiro|
|:-----|:------:|
|Um pouco aberto|7|                                 
|Muito aberto|10|                                   
|Nao muito aberto|3|                                 
|Neutro|5|                                          
|Nao aplicavel, eu nao tenho uma doenca mental|0|    
|Definitivamente nao|0|  

```python
# Mudança dos dados
parametros = {'Um pouco aberto':7,
              'Muito aberto':10,
              'Nao muito aberto':3,
              'Neutro':5,
              'Nao aplicavel, eu nao tenho uma doenca mental':0,
              'Definitivamente nao':0
              }

df_2016['Q8'] = df_2016['Q8'].replace(parametros)

# df_2016.head()
```

# Leitura do PIB e Correlação com DataFrames Anuais


Realizada a leitura do DataFrame com os dados referentes ao PIB. Nos DataFrames referentes aos anos de 2016 até 2020, foi incluída em cada um uma nova coluna referente ao PIB de cada país, correlacionando a coluna Q13 (Qual país você reside?) com a 'Country Name'. No DataFrame final, esta nova coluna foi definida como Q18.

### Leitura do DataFrame PIB

```python
df_pib = pd.read_csv('gdp_data.csv')
```

### 2016

```python
# Seleção das colunas
df_pib_columns_2016 = df_pib[["Country Name", "2016"]]

# Indexando a coluna Country Names
df_pib_2016 = df_pib_columns_2016.set_index('Country Name')

# Correlacionando os DataFrames
df_2016_completo = pd.merge(df_2016, df_pib_2016, left_on='Q13', right_on='Country Name', how='inner')

# Renomeando a nova coluna
df_2016_completo.rename(columns={'2016': 'Q18'}, inplace=True)

# df_2016_completo.head()
```

### 2017

```python
# Seleção das colunas
df_pib_columns_2017 = df_pib[["Country Name", "2017"]]

# Indexando a coluna Country Names
df_pib_2017 = df_pib_columns_2017.set_index('Country Name')

# Correlacionando os DataFrames
df_2017_completo = pd.merge(df_2017, df_pib_2017, left_on='Q13', right_on='Country Name', how='inner')

# Renomeando a nova coluna
df_2017_completo.rename(columns={'2017': 'Q18'}, inplace=True)

# df_2017_completo.head()
```
